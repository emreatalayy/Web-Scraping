```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import time
import sys
from datetime import datetime
from datetime import date
```

```python
def getAndParseURL(url):
    result = requests.get(url,headers={"User-Agent":"Mozilla/5.0"})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
PAGE_NUM = []
def find_page_num():

    html = getAndParseURL("https://www.zingat.com/ortahisar-kiralik-daire")
    for i in range(1,43):
        PAGE_NUM.append("https://www.zingat.com/ortahisar-kiralik-daire" + "?page=" + str(i))
    return PAGE_NUM
find_page_num()
```

```python
ALL_PRODUCT = []
def all_product():
    for i in PAGE_NUM[::]:
        html = getAndParseURL(i)
        for sonuc in html.findAll("a",{"class":"zl-card-inner"}):
            ALL_PRODUCT.append("https://www.zingat.com" + sonuc.get("href"))
    return ALL_PRODUCT
all_product()
```

```python
RESULT = []
for sonuc in ALL_PRODUCT[:872]:

    html = getAndParseURL(sonuc)
    ilan_ismi = html.find("div", {"class":"col-xs-12"}).h1.text
    mahalle = html.find("div", {"class":"detail-location-path"}).h2.text.replace("Ortahisar\n","").replace("Trabzon","").replace(",","").strip()
    fiyat = html.find("div", {"class":"col-md-3 price-info-text"}).find("strong",{"itemprop":"price"}).text.replace("TL","").strip()
    net_m2 = html.find("ul", {"class":"row attribute-detail-list"}).findAll("li",{"class":"col-md-6"})[2].find("span").text.replace("m²","").strip()
    brüt_m2 = html.find("ul", {"class":"row attribute-detail-list"}).findAll("li",{"class":"col-md-6"})[3].find("span").text.replace("m²","").strip()
    oda_salon = html.find("ul", {"class":"row attribute-detail-list"}).findAll("li",{"class":"col-md-6"})[4].find("span").text.strip()
    banyo = html.find("ul", {"class":"row attribute-detail-list"}).findAll("li",{"class":"col-md-6"})[5].find("span").text.strip()
    fotograf = html.find("div", {"class":"detail-images-slide-container hidden-xs hidden-sm"}).a.span.text.replace("+","")
    
    RESULT.append([ilan_ismi,mahalle,fiyat,net_m2,brüt_m2,oda_salon,banyo,fotograf])
    
columns = ["İLAN ADI","MAHALLE", "FİYAT","NET(m²)","BRÜT(m²)","ODA","BANYO SAYISI","FOTOĞRAF SAYISI"]
df = pd.DataFrame.from_records(RESULT, columns=columns)
```
